```python
#print(df)
```

```python
#Problem Statement 1: Delete unnamed columns

#Method 1: Using loc fetch columns where column name does not contact Unnamed
df_without_unnamed_cols = df.loc[:, ~df.columns.str.contains('^Unnamed')]

#Method 2: find the columns by name 'unnamed' then drop those columns.
df_without_unnamed_cols = df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1)

print(df_without_unnamed_cols)
```

```python
#Problem Statement 2: Show the distribution of male and female
print(df["Gender"].describe())
print("\n")
print(df.Gender.value_counts())
```

```python
#Problem Statement 3:Show the top 5 most preferred names
#Method 1: (Python)
print(df_without_unnamed_cols.groupby('Name')['Count'].sum().nlargest(5))

#Method 2: (PandaSQL)
print("\nPandaSQL")
from pandasql import sqldf

pysqldf = lambda statement: sqldf(statement, globals())
new_top_5_names = pysqldf("SELECT Name, Sum(Count) as 'Count' from df_without_unnamed_cols group by Name order by Count desc LIMIT 5;")
print(new_top_5_names)
```

```python
#Problem Statement 4:What is the median name occurence in the dataset

print(df[df.Id == df.median().Id]['Name'])
```

```python
#Problem Statement 5:Distribution of male and female born count by states
#Method 1: (Pandas)
print(df_without_unnamed_cols.groupby(['State','Gender'])['Count'].sum())

#Method 2: (PandaSQL)
print("\nPandaSQL")
from pandasql import sqldf

pysqldf = lambda statement: sqldf(statement, globals())
new_df = pysqldf("SELECT State, Gender, Sum(Count) as 'Count' from df_without_unnamed_cols group by State, Gender;")
print(new_df)
```
